Clasificacion de las caracteristicas y el objetivo de los datos de entrenamiento 

```python
#se realizara un modelo de clasificacion supervisado 
#se separa las caracteristicas de la vivienda del objetivo que es la prediccion del precio
#en la variable dataset_traun_x estaran las caracteristicas de la vivienda de los datos de entrenamiento 
dataset_train_x=dataset_train.drop("SalePrice",axis="columns")
#la variable dataset_train_y tendra almacenado los valores que se quieren predecir de los datos de entrenamiento
dataset_train_y=dataset_train["SalePrice"]
```

```python
print(dataset_train_y)
```

Clasificacion de las caracteristicas y el objetivo de los datos de prueba

```python
#en la variable dataset_test_x se guardan las caracteristocas de los datos de prueba
dataset_test_x=dataset_test.drop("SalePrice",axis="columns")
#en la variable dataset_test_y se guarda el precio que se quiere predecir de los datos de prueba 
dataset_test_y=dataset_test["SalePrice"] 
```

# Implementacion algoritmo XGBoost 

```python
import xgboost as xgb
#realizar la regresion lineal de los datos de entrenamiento
xg_reg=xgb.XGBRegressor(objective ='reg:linear', colsample_bytree =0.85, learning_rate = 0.1,
               max_depth =3,alpha=50, n_estimators =300)
```

```python
xg_reg.fit(dataset_train_x,dataset_train_y)
predicion_xgb=xg_reg.predict(dataset_test_x)
print(predicion_xgb)
```

```python
from sklearn.metrics import mean_squared_error
error_xgb=np.sqrt(mean_squared_error(dataset_test_y,predicion_xgb))
print("El Error que presenta el modelo es: ",error_xgb)
```
